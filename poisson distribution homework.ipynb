```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
number_of_deaths = np.array([0, 1, 2, 3, 4, 5, 6])
probability = np.array([0.5434, 0.3315, 0.1011, 0.0205, 0.0032, 0.0004, 0.0001])
predicted_occurrences = np.array([108.68, 66.3, 20.22, 4.1, 0.64, 0.08, 0.02])
actual_occurrences = np.array([109, 65, 22, 3, 1, 0, 0])

# Create a bar chart
plt.figure(figsize=(10, 6))

bar_width = 0.2
index = np.arange(len(number_of_deaths))

plt.bar(index, probability, bar_width, label='Probability', color='b', align='center')
plt.bar(index + bar_width, predicted_occurrences, bar_width, label='Predicted Occurrences', color='b', align='center')
plt.bar(index + 2*bar_width, actual_occurrences, bar_width, label='Actual Occurrences', color='r', align='center')

plt.xlabel('Number of Deaths')
plt.ylabel('Counts')
plt.title('Bar Chart of Occurrences')
plt.xticks(index + bar_width, number_of_deaths)
plt.legend()

plt.show()
```
